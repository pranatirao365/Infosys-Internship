```python
#load dataset (adjust path if needed)
import os
import pandas as pd
import numpy as np

csv_path = "weather_data.csv"  
print("CSV exists?", os.path.exists(csv_path))

df = pd.read_csv(csv_path, low_memory=False)

```

```python
# 1. Shape and columns
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python

# 2. First few rows
display(df.head())
```

```python

# 3. Info (data types + null counts)
df.info()
```

```python
# 4. Summary of numerical data
display(df.describe())
```

```python

# 5. Missing values
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values:\n", missing)
```

```python
# 6. Duplicates
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Convert last_updated to datetime
df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')

# Ensure temperature is numeric (sometimes string due to formatting issues)
df['temperature_celsius'] = pd.to_numeric(df['temperature_celsius'], errors='coerce')

# Same for PM2.5 (air quality) if needed
df['air_quality_PM2.5'] = pd.to_numeric(df['air_quality_PM2.5'], errors='coerce')

# Re-check data types
df.info()
```

```python
# Fill missing numeric values with median
df['temperature_celsius'].fillna(df['temperature_celsius'].median(), inplace=True)
df['air_quality_PM2.5'].fillna(df['air_quality_PM2.5'].median(), inplace=True)
```

```python
df['condition_text'].fillna(df['condition_text'].mode()[0], inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Check again
print("Duplicates after removal:", df.duplicated().sum())
```

```python
# Define a function to remove outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower, upper)  # replace outliers with boundary values
    return df

# Apply for temperature and PM2.5
df = remove_outliers(df, 'temperature_celsius')
df = remove_outliers(df, 'air_quality_PM2.5')

# Quick check
df[['temperature_celsius', 'air_quality_PM2.5']].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots look nicer
sns.set(style="whitegrid")
%matplotlib inline

```

```python
# List of important numeric columns for EDA
numeric_cols = ['temperature_celsius', 'air_quality_PM2.5', 'humidity', 'wind_kph', 'pressure_mb', 'uv_index']

# Histograms
df[numeric_cols].hist(figsize=(12,8), bins=30, color='skyblue')
plt.suptitle("Distribution of Key Numeric Features", fontsize=16)
plt.show()
```

```python
# Compute correlation matrix
corr = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Key Features")
plt.show()
```

```python
# Resample by day (mean PM2.5 per day)
df_daily = df.set_index('last_updated').resample('D')['air_quality_PM2.5'].mean()

plt.figure(figsize=(12,5))
df_daily.plot(color='orange')
plt.title("Daily Average PM2.5 over Time")
plt.ylabel("PM2.5")
plt.xlabel("Date")
plt.show()
```
